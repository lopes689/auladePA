```python
# Escrever um programa que verifica se o número digitado é par ou ímpar

# Entrada

numero = int (input("Digite um número: "))

# Processamento / Saída

if numero % 2 == 0:

  print (f" O número {numero} é par.")

else:
  print (f"O número {numero} é impar")
```

```python
# Escrever um programa que o usuário digita o nome e a idade e verifica se o usuário é criança, adulto ou adolescente

# Entrada

nome = str (input("Digite o seu nome: "))
idade = int (input("Digite a sua idade: "))

# Processamento / Saída

if idade < 12 :
  print(f"{nome}, voce é uma criança")

elif idade < 18 :
  print (f"{nome}, voce é um Adolescente")

else:

  print (f"{nome}, voce é um Adulto")
```
