```python
# 1 - Contagem simples, o usuário digita um número e o programa imprimi do zero até o número.

numero = int (input("digite um numero"))

for i in range (numero + i):
  print (i)
```

```python
# 2 - Contagem regressiva, o usuário um número e o programa imprimi do número até o zero.

numero = int(input("digite um numero: "))

for i in range(numero + i):
  print(numero - i)
```

```python
# 3 - Números pares, o usuário digita um número e o programa imprimi os números pares.

numero = int(input("digite um numero: "))

for i in range(numero + 1):
  if i % 2==0:
    print (i)
```

```python
# 4 - Números impares, o usuário digita um número e o programa imprimi os números ímpares.

numero = int(input("digite um numero: "))

for i in range(numero + 1):
  if i % 2==1:
    print (i)
```

```python
# 5 - Soma, o usuário digita um número e o programa soma os números do zero até o número.

numero = int(input("digite um numero: "))

soma = 0
for i in range(numero + 1):
  soma += 1
print(soma)
```

```python
# 6 - Tabuada, o usuário digita um número e o programa imprimi a tabuada do número.

numero = int (input("digite um numero: "))

if 1 <= numero <=10:
  print(f"tabuada do {numero}:")
  for i in range(1,11):
    print( f"{numero} x {i} = {numero * i}")
```
