```python
# Solicite ao usuário um número inteiro positivo.

# Utilizando um laço de repetição, calcule e exiba a soma de
# todos os números de 1 até o número informado.

# Exemplo:

# Digite um número: 5

# Resultado: 15

# Pois:

# 1 + 2 + 3 + 4 + 5 = 15
a = int(input("digite um numero "))
for i in range(1,a):
    a += i
print(a)
```

```python
# Solicite ao usuário um número inteiro positivo e calcule
# seu fatorial utilizando um laço de repetição.

# Exemplo:

# Digite um número: 5

# Fatorial = 120

# Pois:

# 5! = 5 × 4 × 3 × 2 × 1

# Observação:
# O programa deve validar se o número informado é positivo.

a = int(input("digite um numero "))
try:
  for i in range(1,a):
    a = a * i
  print(a )
except a >= 0:
  print("digite um numero positivo")
```

```python
# Um caixa eletrônico possui notas de:

# R$ 100
# R$ 50
# R$ 20
# R$ 10
# R$ 5
# R$ 2
# R$ 1

# Solicite ao usuário o valor do saque e informe a menor
# quantidade possível de notas para realizar o pagamento.

# Exemplo:

# Digite o valor do saque: 287

# Saída esperada:

# 2 nota(s) de R$100
# 1 nota(s) de R$50
# 1 nota(s) de R$20
# 1 nota(s) de R$10
# 1 nota(s) de R$5
# 1 nota(s) de R$2

# Requisitos:
# - Utilizar laços de repetição.
# - Não utilizar listas.
# - Exibir apenas as notas utilizadas.

# Solicita o valor do saque ao usuário
valor = int(input("Digite o valor do saque: "))

# Define a maior cédula disponível para iniciar o ciclo
nota_atual = 100

# Laço de repetição principal para processar o valor restante
while valor > 0:
    quantidade_notas = 0

    # Conta quantas notas da cédula atual cabem no valor do saque
    while valor >= nota_atual:
        valor -= nota_atual
        quantidade_notas += 1

    # Exibe apenas se a nota foi utilizada
    if quantidade_notas > 0:
        print(f"{quantidade_notas} nota(s) de R${nota_atual}")

    # Atualiza o valor da nota para a próxima denominação sem usar listas
    if nota_atual == 100:
        nota_atual = 50
    elif nota_atual == 50:
        nota_atual = 20
    elif nota_atual == 20:
        nota_atual = 10
    elif nota_atual == 10:
        nota_atual = 5
    elif nota_atual == 5:
        nota_atual = 2
    elif nota_atual == 2:
        nota_atual = 1
```
